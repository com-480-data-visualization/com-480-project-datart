```python
def apply_region (row):
    if (row['Code département'] in ['67','68']):
        val = 'Grand Est'
    elif (row['Code département'] in ['24','33','40','47','64']):
        val = 'Nouvelle-Aquitaine'
    elif (row['Code département'] in ['3','15','43','63']):
        val = 'Auvergne-Rhône-Alpes'
    elif (row['Code département'] in ['14','50','61']):
        val = 'Normandie'
    elif (row['Code département'] in ['21','58','71','89']):
        val = 'Bourgogne-Franche-Comté'
    elif (row['Code département'] in ['22','29','35','56']):
        val = 'Bretagne'
    elif (row['Code département'] in ['18','24','36','37','41','45']):
        val = 'Centre-Val-de-Loire'
    elif (row['Code département'] in ['8','10','51','52']):
        val = 'Grand Est'
    elif (row['Code département'] in ['2A','2B']):
        val = 'Corse'
    elif (row['Code département'] in ['25','39','70','90']):
        val = 'Bourgogne-Franche-Comté'
    elif (row['Code département'] in ['27','76']):
        val = 'Normandie'
    elif (row['Code département'] in ['75','77','78','91','92','93','94','95']):
        val = 'Île-de-France'
    elif (row['Code département'] in ['11','30','34','48','66']):
        val = 'Occitanie'
    elif (row['Code département'] in ['19','23','87']):
        val = 'Nouvelle-Aquitaine'
    elif (row['Code département'] in ['54','55','57','88']):
        val = 'Grand Est'
    elif (row['Code département'] in ['9','12','31','32','46','65','81','82']):
        val = 'Occitanie'
    elif (row['Code département'] in ['44','49','53','72','85']):
        val = 'Pays de la Loire'
    elif (row['Code département'] in ['2','60','80']):
        val = 'Hauts-de-France'
    elif (row['Code département'] in ['59','62']):
        val = 'Hauts-de-France'
    elif (row['Code département'] in ['16','17','79','86']):
        val = 'Nouvelle-Aquitaine'
    elif (row['Code département'] in ['4','5','6','13','83','84']):
        val = "Provence-Alpes-Côte d'Azur"
    elif (row['Code département'] in ['1','7','26','38','42','69','73','74']):
        val = 'Auvergne-Rhône-Alpes'
    elif (row['Code département'] in ['971']):
        val = 'Guadeloupe '
    elif (row['Code département'] in ['972']):
        val = 'Martinique '
    elif (row['Code département'] in ['973']):
        val = 'Guyane'
    elif (row['Code département'] in ['974']):
        val = 'La Réunion'
    elif (row['Code département'] in ['976']):
        val = 'Mayotte'
    elif (row['Code département'] in ['984','986','987','988','975']):
        val = 'DOM'
    else :
        val = 'no region'
    return val
```

```python
df['region'] = df.apply(apply_region, axis=1)
```
